```python
import os
os.chdir("..") 

import sys
sys.path.append(os.path.abspath(".."))

import pandas as pd
import matplotlib.pyplot as plt
from run import main  
import os

device = 'cuda:2'
num_rounds = 500
batch_size = 64
lr = 5e-3
c = 0.1
local_epoch = 10
```

### Ablation Study on how different level of added noise will affect model performances

```python
noise_level_list = [5e-5, 1e-4, 2e-4, 5e-4, 1e-3, 2e-3]
results_noise = []
```

```python
for noise_level in noise_level_list:
    print(f"Running FedAvg: Noice Level={noise_level}...")
    test_acc, training_acc = main(
        num_rounds=num_rounds,
        c=c,
        local_epochs=local_epoch,
        batch_size=batch_size,
        lr=lr,
        device=device,
        noise_level=noise_level
    )
    results_noise.append({"noise_level": noise_level, 
                           "test_accuracy": test_acc,
                           "training_accuracy": training_acc})

# Plot test accuracy vs noise_level
df_noise = pd.DataFrame(results_noise)
plt.figure(figsize=(6,4))
plt.plot(df_noise["noise_level"], df_noise["test_accuracy"], 
         marker='o', label='Test Accuracy', color='blue')
plt.plot(df_noise["noise_level"], df_noise["training_accuracy"], 
         marker='s', label='Training Accuracy', color='orange', linestyle='--')
plt.xscale('log')
plt.xlabel("Noise Levels")
plt.ylabel("Accuracy")
plt.title(f"FedAvg Accuracy vs Noise Level (C={c}, Local Epochs={local_epoch}, Num_Rounds={num_rounds}, Batch Size={batch_size}, LR={lr})")
plt.grid(True)
plt.legend()
plt.show()
```

### Training Curve for each round of experiment

```python
fig, axes = plt.subplots(3, 2, figsize=(12, 16))
axes = axes.flatten()

for idx, noise_level in enumerate(noise_level_list):

    file_path = f"runs/noise={noise_level}_{num_rounds}rounds_{c}C_{local_epoch}epochs_{batch_size}bs_{lr}lr.png"

    ax = axes[idx]
    if os.path.exists(file_path):
        img = plt.imread(file_path)
        ax.imshow(img)
        ax.set_title(f"Noise = {noise_level}")
        ax.axis("off")

fig.suptitle(
    f"Training Curves in Part 2\n"
    f"{num_rounds} Rounds | C={c} | Local Epochs={local_epoch} | Batch={batch_size} | LR={lr}",
    fontsize=16,
    y=1.02
)

plt.tight_layout()
plt.show()
```
