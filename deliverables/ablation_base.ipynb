```python
import os
os.chdir("..") 

import sys
sys.path.append(os.path.abspath(".."))

import pandas as pd
import matplotlib.pyplot as plt
from run import main  

device = 'cuda:1'
num_rounds = 800
batch_size = 64
lr = 5e-3
```

### Experiment 1: Effect of # of local epochs

```python
c_fixed = 0.1 
local_epochs_list = [1, 5, 10, 20]
results_epochs = []
```

```python
for local_epochs in local_epochs_list:
    print(f"Running FedAvg: C={c_fixed}, local_epochs={local_epochs} ...")
    test_acc = main(
        num_rounds=num_rounds,
        c=c_fixed,
        local_epochs=local_epochs,
        batch_size=batch_size,
        lr=lr,
        device=device
    )
    results_epochs.append({"local_epochs": local_epochs, "test_accuracy": test_acc})

# Plot test accuracy vs local_epochs
df_epochs = pd.DataFrame(results_epochs)
plt.figure(figsize=(6,4))
plt.plot(df_epochs["local_epochs"], df_epochs["test_accuracy"], marker='o')
plt.xlabel("Local Epochs")
plt.ylabel("Test Accuracy")
plt.title(f"FedAvg Test Accuracy vs Local Epochs (C={c_fixed}, Num_Rounds={num_rounds}, Batch Size={batch_size}, LR={lr})")
plt.grid(True)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for idx, local_epochs in enumerate(local_epochs_list):

    file_path = f"runs/{num_rounds}rounds_{c_fixed}C_{local_epochs}epochs_{batch_size}bs_{lr}lr.png"

    ax = axes[idx]
    if os.path.exists(file_path):
        img = plt.imread(file_path)
        ax.imshow(img)
        ax.set_title(f"{local_epochs} Local Epochs")
        ax.axis("off")
    else:
        ax.text(0.5, 0.5, "Image not found", ha="center", va="center")
        ax.set_title(f"{local_epochs} Local Epochs")
        ax.axis("off")

fig.suptitle(
    f"Training Curves in Experiment 1\n"
    f"{num_rounds} Rounds | C={c_fixed} | Batch={batch_size} | LR={lr}",
    fontsize=16,
    y=1.02
)

plt.tight_layout()
plt.show()
```

### Experiment 2: Effect of c

```python
local_epochs_fixed = 10
c_list = [0.01, 0.04, 0.07, 0.1]
results_c = []
```

```python
for c_percentage in c_list:
    print(f"Running FedAvg: C={c_percentage}, local_epochs={local_epochs_fixed} ...")
    test_acc = main(
        num_rounds=num_rounds,
        c=c_percentage,
        local_epochs=local_epochs_fixed,
        batch_size=batch_size,
        lr=lr,
        device=device
    )
    results_c.append({"C_percentage": c_percentage, "test_accuracy": test_acc})

# Plot test accuracy vs C
df_c = pd.DataFrame(results_c)
plt.figure(figsize=(6,4))
plt.plot(df_c["C_percentage"], df_c["test_accuracy"], marker='o')
plt.xlabel("C (Fraction of Clients)")
plt.ylabel("Test Accuracy")
plt.title(f"FedAvg Test Accuracy vs C (Local Epochs={local_epochs_fixed}, Num_Rounds={num_rounds}, Batch Size={batch_size}, LR={lr})")
plt.grid(True)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for idx, c in enumerate(c_list):

    file_path = f"runs/{num_rounds}rounds_{c}C_{local_epochs_fixed}epochs_{batch_size}bs_{lr}lr.png"

    ax = axes[idx]
    if os.path.exists(file_path):
        img = plt.imread(file_path)
        ax.imshow(img)
        ax.set_title(f"# of Client = {int(c*100)}")
        ax.axis("off")
    else:
        ax.text(0.5, 0.5, "Image not found", ha="center", va="center")
        ax.set_title(f"C = {c}")
        ax.axis("off")

fig.suptitle(
    f"Training Curves for Experiment 2\n"
    f"{num_rounds} Rounds | Local Epoch={local_epochs_fixed} | Batch={batch_size} | LR={lr}",
    fontsize=16,
    y=1.02
)

plt.tight_layout()
plt.show()
```
